```python
import random
guess=""
i=1
win=0
while 1:
    guess=input("擲骰子猜大小，1~3為小(輸入小/s)，4~6為大(輸入大/b)，(輸入1停止): ")
    if(guess=="1"):
        break
    dice=random.randint(1,6)
    print("第"+str(i)+"次結果:"+str(dice))
    if((dice<=3 and (guess=="小" or guess=="s"))or(dice>=4 and (guess=="大" or guess=="b"))):
        win+=1
        print("猜對了！勝率:"+str(int(win/i*100))+"% ",str(win)+"/"+str(i))
    else:
        print("猜錯了！勝率:"+str(int(win/i*100))+"% ",str(win)+"/"+str(i))
    i+=1
print("-----程式終止-----")
```

```python
import random
guess=""
i=1
win=0
while 1:
    times=input("擲幾次骰子？，(輸入0停止): ")
    if(times=="0"):
        break
    guess=input("擲"+times+"次骰子猜大小，1~3為小(輸入小/s)，4~6為大(輸入大/b)，(輸入0停止): ")
    if(guess=="0"):
        break
    for j in range(0,int(times)):
        dice=random.randint(1,6)
        print("第"+str(i)+"次結果:"+str(dice))
        if((dice<=3 and (guess=="小" or guess=="s"))or(dice>=4 and (guess=="大" or guess=="b"))):
            win+=1
        i+=1
    print("勝率:"+str(round(win/(i-1)*100,2))+"% ",str(win)+"/"+str(i-1))
print("-----程式終止-----")
```
