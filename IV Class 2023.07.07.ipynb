```python
import jieba
text="""元素反应是通过对目标（敌人、玩家或物体）应用某些元素效果组合来触发的。

当一种元素应用于目标时，它会产生一种元素光环，该光环会随着时间的推移而衰减。当第二个元素应用于具有现有元素光环的目标时，就会发生元素反应。

触发元素反应将消耗现有的光环，消耗量取决于触发的反应类型和触发元素的计量单位。这将缩短现有光环的持续时间或完全消除它。

请注意，某些技能在重新应用其元素效果之前有内部冷却时间，因此可能无法在每次攻击时触发元素反应。

元素规范理论（一种理解元素相互作用的框架）深入解释了元素反应的某些更复杂的力学。"""

cut=jieba.lcut(text)
count={}
for i in cut:
    if i in count:
        count[i]+=1
    else:
        count[i]=1
maximum=0
for i in count:
    if(count[i]>maximum):
        maximum=count[i]
reorder={}
while(maximum!=0):
    for i in count:
        if(count[i]==maximum):
            reorder[i]=count[i]
    maximum-=1

print(count)
print(reorder)
```

```python
import requests
import json
import time
import jieba
query=input("輸入Access Token: ")
print()
print("Loading",end="")
req=requests.get("https://graph.facebook.com/v17.0/me/posts?access_token="+query)
translate=json.loads(req.text)
combination=[]
while "paging" in translate:
    post=1
    for i in translate["data"]:
        if "message" in i:
            cut=jieba.lcut(i["message"])
            combination=combination+cut
        post+=1
        time.sleep(0.3)
        if(post%4==0 and post<=12):
            print(".",end="")
        elif(post%4==0):
            print("\b",end="")
    nexturl=translate["paging"]["next"]
    req=requests.get(nexturl)
    translate=json.loads(req.text)
count={}
for i in combination:
    if i in count:
        count[i]+=1
    else:
        count[i]=1
maximum=0
for i in count:
    if(count[i]>maximum):
        maximum=count[i]
reorder={}
while(maximum!=0):
    for i in count:
        if(count[i]==maximum):
            reorder[i]=count[i]
    maximum-=1
print("\n")
for i in reorder:
    print(i,reorder[i])
```
