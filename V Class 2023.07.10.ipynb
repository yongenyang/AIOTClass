```python
from bs4 import BeautifulSoup
path="web.html"
with open(path,"r",encoding="UTF-8") as f:
    soup=BeautifulSoup(f.read(),"html")
print(soup.prettify())
print("\n檢索第一個名為a的tag:",soup.a.text,sep="\n")
print("\n檢索第一個名為h1,屬性為abc的tag:",soup.find("h1",class_="abc"),sep="\n")
print("\n檢索所有名為h1,屬性為abc的tag,輸出清單:",soup.find_all("h1",class_="abc"),sep="\n")
```

```python
import requests
from bs4 import BeautifulSoup
url104=requests.get("https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=%E5%A4%A7%E6%95%B8%E6%93%9A%E5%B7%A5%E7%A8%8B%E5%B8%AB&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=15&asc=0&page=1&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1")
soup104=BeautifulSoup(url104.text)
count=1
page=1
while(page<21):
    for i in soup104.select("article[data-qa-id=jobSeachResult]"):
        print("第",count,"個職缺：")
        print("職稱： "+i("a")[0].text)
        print("公司： "+i("a")[1].text[:-20],end="")
        print("地區： "+i("ul")[1]("li")[0].text)
        if((i("a",class_="b-tag--default"))!=[]):
            print("薪資： ",end="")
            if(i("a",class_="b-tag--default")[0].text[1]=="薪"):
                print(i("a",class_="b-tag--default")[0].text)
            else:
                print(i("span",class_="b-tag--default")[0].text)
        elif(i("span",class_="b-tag--default")):
            print("薪資： "+i("span",class_="b-tag--default")[0].text)
        print("網址： https:"+i("a")[0]["href"])
        print("")
        count+=1
    page+=1
    url104=requests.get("https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=%E5%A4%A7%E6%95%B8%E6%93%9A%E5%B7%A5%E7%A8%8B%E5%B8%AB&expansionType=area%2Cspec%2Ccom%2Cjob%2Cwf%2Cwktm&order=15&asc=0&page="+str(page)+"&mode=s&jobsource=2018indexpoc&langFlag=0&langStatus=0&recommendJob=1&hotJob=1")
    soup104=BeautifulSoup(url104.text)
```
